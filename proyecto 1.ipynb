```python
import os
from datetime import datetime

historial = []
RUTA_ARCHIVO = "datos/historial.txt"
```

```python
def validar_numero(mensaje):
    while True:
        try:
            return float(input(mensaje))
        except ValueError:
            print("Error: Ingresa un número válido.")
```

```python
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        return "Error: No se puede dividir entre cero"
    return a / b

def modulo(a, b):
    return a % b

def potencia(a, b):
    return a ** b
```

```python
def bytes_a_kb(b): return b / 1024
def kb_a_bytes(kb): return kb * 1024
def kb_a_mb(kb): return kb / 1024
def mb_a_kb(mb): return mb * 1024
def mb_a_gb(mb): return mb / 1024
def gb_a_mb(gb): return gb * 1024
```

```python
def decimal_a_binario(n):
    n = int(n)
    if n == 0:
        return "0"
    binario = ""
    while n > 0:
        residuo = n % 2
        binario = str(residuo) + binario
        n = n // 2
    return binario
```

```python
def decimal_a_hexadecimal(n):
    return hex(int(n))[2:]

def binario_a_decimal(b):
    return int(b, 2)

def hexadecimal_a_decimal(h):
    return int(h, 16)
```

```python
def agregar_historial(operacion, num1, num2, resultado):
    global historial
    
    fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entrada = f"{fecha} | {operacion} | {num1}, {num2} = {resultado}"
    
    historial.append(entrada)
    
    if len(historial) > 10:
        historial.pop(0)
```

```python
def ver_historial():
    if not historial:
        print(" Historial vacío.")
    else:
        print("\n===== HISTORIAL =====")
        for h in historial:
            print(h)

def limpiar_historial():
    global historial
    historial = []
    print(" Historial limpiado.")
```

```python
def guardar_historial():
    if not os.path.exists("datos"):
        os.makedirs("datos")
        
    with open(RUTA_ARCHIVO, "w") as archivo:
        for h in historial:
            archivo.write(h + "\n")

def cargar_historial():
    global historial
    
    if os.path.exists(RUTA_ARCHIVO):
        with open(RUTA_ARCHIVO, "r") as archivo:
            historial = archivo.read().splitlines()
```

```python
def menu():
    cargar_historial()
    
    while True:
        print("\n========= CALCULADORA DIGITAL =========")
        print("1. Calculadora Básica")
        print("2. Conversor de Datos")
        print("3. Sistemas Numéricos")
        print("4. Ver Historial")
        print("5. Limpiar Historial")
        print("6. Salir")
        
        opcion = input("Selecciona una opción: ")
        
        # CALCULADORA BÁSICA
        if opcion == "1":
            a = validar_numero("Número 1: ")
            b = validar_numero("Número 2: ")
            
            print("1.Suma 2.Resta 3.Multiplicación 4.División 5.Módulo 6.Potencia")
            op = input("Operación: ")
            
            if op == "1":
                r = sumar(a,b)
                agregar_historial("Suma",a,b,r)
                print("Resultado:", r)
                
            elif op == "2":
                r = restar(a,b)
                agregar_historial("Resta",a,b,r)
                print("Resultado:", r)
                
            elif op == "3":
                r = multiplicar(a,b)
                agregar_historial("Multiplicación",a,b,r)
                print("Resultado:", r)
                
            elif op == "4":
                r = dividir(a,b)
                agregar_historial("División",a,b,r)
                print("Resultado:", r)
                
            elif op == "5":
                r = modulo(a,b)
                agregar_historial("Módulo",a,b,r)
                print("Resultado:", r)
                
            elif op == "6":
                r = potencia(a,b)
                agregar_historial("Potencia",a,b,r)
                print("Resultado:", r)
                
            else:
                print(" Operación inválida")
        
        # CONVERSOR
        elif opcion == "2":
            valor = validar_numero("Valor: ")
            print("1.Bytes→KB 2.KB→Bytes 3.KB→MB 4.MB→KB 5.MB→GB 6.GB→MB")
            conv = input("Conversión: ")
            
            if conv == "1":
                print("Resultado:", bytes_a_kb(valor))
            elif conv == "2":
                print("Resultado:", kb_a_bytes(valor))
            elif conv == "3":
                print("Resultado:", kb_a_mb(valor))
            elif conv == "4":
                print("Resultado:", mb_a_kb(valor))
            elif conv == "5":
                print("Resultado:", mb_a_gb(valor))
            elif conv == "6":
                print("Resultado:", gb_a_mb(valor))
            else:
                print(" Conversión inválida")
        
        # SISTEMAS NUMÉRICOS
        elif opcion == "3":
            print("1.Decimal→Binario")
            print("2.Decimal→Hexadecimal")
            print("3.Binario→Decimal")
            print("4.Hexadecimal→Decimal")
            
            sn = input("Opción: ")
            
            if sn == "1":
                n = validar_numero("Decimal: ")
                print("Resultado:", decimal_a_binario(n))
                
            elif sn == "2":
                n = validar_numero("Decimal: ")
                print("Resultado:", decimal_a_hexadecimal(n))
                
            elif sn == "3":
                b = input("Binario: ")
                print("Resultado:", binario_a_decimal(b))
                
            elif sn == "4":
                h = input("Hexadecimal: ")
                print("Resultado:", hexadecimal_a_decimal(h))
                
            else:
                print(" Opción inválida")
        
        elif opcion == "4":
            ver_historial()
            
        elif opcion == "5":
            limpiar_historial()
            
        elif opcion == "6":
            guardar_historial()
            print(" Historial guardado. Programa finalizado.")
            break
            
        else:
            print(" Opción inválida.")
```

```python
menu ()
```
